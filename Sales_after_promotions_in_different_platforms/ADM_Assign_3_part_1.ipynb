```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

```python
df = pd.read_csv("/test5.csv")
df
```

```python
df.plot.line(x='week', y=['tv_spend','tv_Sales_2', 'base_sales'], figsize =(13.5,13.5))
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '-', linewidth = '.5', color = 'r')
#plt.grid(which = 'minor', linestyle = '-', linewidth = '.5', color = 'red')
plt.grid(which = 'minor', linestyle = '-', linewidth = '.2', color = 'b')
plt.title('TV AD SALES vs. BASE SALES', fontsize=30)
plt.savefig('TV_ad_Sales.png')

files.download("TV_ad_Sales.png") 
```
